```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_csv("test.csv", header = 0)
```

```python
df.head()
```

```python
def convert(t):
    t_convert = None
    try: 
        ix = t.index('.')
        t = t[:ix]
    except Exception as e:
        print(e)
    
    try:
        t_convert = datetime.strptime(t, "%H:%M:%S")
    except Exception as e:
        print(f"{e}, treating it as a %M:%S situation")
        t_convert = datetime.strptime(t,"%M:%S")
        
    if t_convert != None:
        return round(t_convert.hour*60 + t_convert.minute + t_convert.second/60,1)
    else:
        return "NA"
        
```

```python
convert("26:53.1")
```

```python
df["Alloted Time"] = df.apply(lambda row: convert(row["Alloted Time"]),axis = 1)
```
